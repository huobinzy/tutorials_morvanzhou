```python
import numpy as np
d1 = np.array(['A','1','%'])
print(d1)
d2 = np.arange(0,100,10,dtype=int)
print(d2)
d3 = np.arange(0,100,5).reshape(4,5)
print(d3)
print('\n','dim:',d3.ndim,'\n','shape:',d3.shape,'\n','size:',d3.size)
d4 = np.zeros((3,4))
print(d4)
d5 = np.empty((5,8))
print(d5)
```

```python
import numpy as np
d6 = np.linspace(0,100,20).reshape(5,4)
print(d6)
print(d6[1][2])
print(d6[1,2])
print(d6[:,2])
print(d6[2,:])
print(d6[1:3,:])
for row in d6:
    print(row)
for col in d6.T:
    print(col)
print(d6.flatten())
for item in d6.flat:
    print(item)
```

```python
import numpy as np
a = np.array([1,2,3])
b = np.array([4,5,6])
c = np.vstack((a,b))
d = np.hstack((a,b))
e = np.concatenate((a,b,b,a),axis=0)
print(a)
print(a.shape)
print(b)
print(c)
print(d)
print(e)
print('\n')
a2 = a[:,np.newaxis]
b2 = b[:,np.newaxis]
c2 = np.vstack((a2,b2))
d2 = np.hstack((a2,b2))
print(a2.shape)
print(b2.shape)
print(c2)
print(d2)
```

```python
import numpy as np
d6 = np.linspace(0,100,20).reshape(5,4)
# 等尺寸分割
print(np.split(d6,2,axis=1))
# 不等尺寸分割
print(np.array_split(d6,3,axis=1))
# 纵向等尺寸分割
print(np.vsplit(d6,5))
# 横向等尺寸分割
print(np.hsplit(d6,2))
```
